```python
import time
```

```python
languages=['Python',
 'HTML',
 'CSS',
 'R',
 'Go',
 'JavaScript',
 'C++',
 'TypeScript',
 'Ruby',
 'Scala',
 'Julia',
 'Java',
 'Haskell',
 'C',
'C++',
'C#',
'Swift',
'SQL',
'PHP',
'Perl',
'Kotlin',
'Rust',
'Scheme',
'Erlang']
```

```python
dataset=[]
for year in range(2010,2020):
    for language in languages:
        succeed=False
        while not succeed:
            url = f'https://api.github.com/search/repositories?q= created:{year}-01-01..{year}-12-31 language:{language.lower()}'
            print(url)
            r = requests.get(url)
            json_pydata = r.json()
            try:
                num_repo = json_pydata['total_count']
                print(f'{year} {language}: {num_repo}')
                dataset.append({'year':year,'language':language,'num':num_repo})
                succeed=True
            except Exception as e:
                print(f'Error: {e}')
                print(json_pydata)
                succeed=False
                time.sleep(20) 
```
